## ‚Ññ10 –ò–º–µ—é—Ç—Å—è –¥–≤–∞  ùëõ ‚Äì–º–µ—Ä–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞  ùë•  –∏  ùë¶ , –∫–æ—Ç–æ—Ä—ã–µ –∑–∞–¥–∞—é—Ç –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã  ùëõ  —Ç–æ—á–µ–∫ –Ω–∞ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ (—Å–ª—É—á–∞–π–Ω—ã–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞). –ù–∞–π—Ç–∏ –Ω–∞–∏–±–æ–ª–µ–µ –±–ª–∏–∑–∫–∏–µ –¥—Ä—É–≥ –¥—Ä—É–≥—É —Ç–æ—á–∫–∏.

````python
import numpy as np

n = int(input("–í–≤–µ–¥–∏—Ç–µ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –≤–µ–∫—Ç–æ—Ä–æ–≤ (n): "))
num_points = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ—á–µ–∫: "))

x = np.random.randint(0, 100, size=(num_points, n))
y = np.random.randint(0, 100, size=(num_points, n))

def euclidean_distance(point1, point2):
    return np.linalg.norm(point1 - point2)

min_distance = float('inf')
closest_points = None

for i in range(num_points):
    for j in range(num_points):
        distance = euclidean_distance(x[i], y[j])
        if distance < min_distance:
            min_distance = distance
            closest_points = (x[i], y[j])

print("–ù–∞–∏–±–æ–ª–µ–µ –±–ª–∏–∑–∫–∏–µ —Ç–æ—á–∫–∏:")
print("–¢–æ—á–∫–∞ x:", closest_points[0])
print("–¢–æ—á–∫–∞ y:", closest_points[1])
print("–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ:", min_distance)
````

````python
import math

x1 = float(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É x –Ω–∞—á–∞–ª–∞ –≤–µ–∫—Ç–æ—Ä–∞ x: "))
y1 = float(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É y –Ω–∞—á–∞–ª–∞ –≤–µ–∫—Ç–æ—Ä–∞ x: "))
x2 = float(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É x –∫–æ–Ω—Ü–∞ –≤–µ–∫—Ç–æ—Ä–∞ x: "))
y2 = float(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É y –∫–æ–Ω—Ü–∞ –≤–µ–∫—Ç–æ—Ä–∞ x: "))
x3 = float(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É x –Ω–∞—á–∞–ª–∞ –≤–µ–∫—Ç–æ—Ä–∞ y: "))
y3 = float(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É y –Ω–∞—á–∞–ª–∞ –≤–µ–∫—Ç–æ—Ä–∞ y: "))
x4 = float(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É x –∫–æ–Ω—Ü–∞ –≤–µ–∫—Ç–æ—Ä–∞ y: "))
y4 = float(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É y –∫–æ–Ω—Ü–∞ –≤–µ–∫—Ç–æ—Ä–∞ y: "))

d = math.sqrt((x2-x1)**2+(y2-y1)**2)
print(f"–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ: {d}")
````

## ‚Ññ11–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –∑–∞–¥–∞–Ω –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏ —Å–≤–æ–∏—Ö –≤–µ—Ä—à–∏–Ω:  (ùë•1,ùë¶1),(ùë•2,ùë¶2),(ùë•3,ùë¶3) . –ó–Ω–∞—á–µ–Ω–∏—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Å –ø–æ–º–æ—â—å—é –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏—è. –û–Ω–∏ –º–æ–≥—É—Ç –±—ã—Ç—å –Ω–µ—Ü–µ–ª—ã–º–∏. –ù–∞–π—Ç–∏ –ø–µ—Ä–∏–º–µ—Ç—Ä –∏ –ø–ª–æ—â–∞–¥—å —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞.


````python
import math

x1, y1 = map(float, input("–í–≤–µ–¥–∏—Ç–µ x1 –∏ y1 —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª: ").split())
x2, y2 = map(float, input("–í–≤–µ–¥–∏—Ç–µ x2 –∏ y2 —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª: ").split())
x3, y3 = map(float, input("–í–≤–µ–¥–∏—Ç–µ x3 –∏ y3 —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª: ").split())

side1 = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
side2 = math.sqrt((x3 - x2)**2 + (y3 - y2)**2)
side3 = math.sqrt((x1 - x3)**2 + (y1 - y3)**2)

P = side1 + side2 + side3
s = P / 2
area = math.sqrt(s*(s-side1)*(s-side2)*(s-side3))

print(f"–ü–µ—Ä–∏–º–µ—Ç—Ä : {P:.3f}")
print(f"–ü–ª–æ—â–∞–¥—å : {area:.3f}")
````

## ‚Ññ12 –ü—É—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤–æ–¥–∏—Ç 5 —á–∏—Å–µ–ª: –∞, b, c, d, e. –†–µ–∞–ª–∏–∑—É–π—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É —Ä–∞—Å—á–µ—Ç–∞ –≤—ã—Ä–∞–∂–µ–Ω–∏—è –≤–∏–¥–∞  ùëüùëíùë†=ùëé+ùëèùëê2ùëë‚àíùëí+ùëÇùë†ùë°[ùëèùëí/ùëê] , –≥–¥–µ  ùëÇùë†ùë°  - –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è. –£—á–µ—Å—Ç—å –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –¥–µ–ª–µ–Ω–∏—è –Ω–∞ 0.

````python
import math
a = int(input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ a: "))
b = int(input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ b: "))
c = int(input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ c: "))
d = int(input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ d: "))
e = int(input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ e: "))

if c == 0:
    print("–î–µ–ª–µ–Ω–∏–µ –Ω–∞ 0 –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ")
else:
    fraction_result = b * e / c
    remainder = fraction_result % 1
    result = a + (b * c) / (2 * d) - e + remainder
    print (f"–†–µ–ª—É–∑–ª—å—Ç–∞—Ç : {result:.3f}")
````

### –µ—Å–ª–∏ —Å–∏–º–≤–æ–ª * –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –≤ —Å—Ç—Ä–æ–∫–µ —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Ä–∞–∑, –≤—ã–≤–µ–¥–∏—Ç–µ –µ–≥–æ –∏–Ω–¥–µ–∫—Å;
### –µ—Å–ª–∏ —Å–∏–º–≤–æ–ª * –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –¥–≤–∞ –∏ –±–æ–ª–µ–µ —Ä–∞–∑, –≤—ã–≤–µ–¥–∏—Ç–µ –∏–Ω–¥–µ–∫—Å –µ–≥–æ –ø–µ—Ä–≤–æ–≥–æ, –≤—Ç–æ—Ä–æ–≥–æ –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è, —É–¥–∞–ª–∏—Ç–µ –ø–µ—Ä–≤—ã–π –∏ –ø–æ—Å–ª–µ–¥–Ω–∏–π —Å–∏–º–≤–æ–ª * –∏–∑ —Å—Ç—Ä–æ–∫–∏.

````python
string = input("–í–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫—É:  ")

if '*' not in string:
    print("–Ω–µ—Ç —Å–∏–º–≤–æ–ª–∞")
else:
    first_occurrence = string.find('*')
    last_occurrence = string.rfind('*')

    if first_occurrence == last_occurrence:
        print("–ò–Ω–¥–µ–∫—Å —Å–∏–º–≤–æ–ª–∞ *:", first_occurrence)
    else:
        second_occurrence = string.find('*', first_occurrence + 1)
        print("–ò–Ω–¥–µ–∫—Å—ã —Å–∏–º–≤–æ–ª–∞ *:", first_occurrence, second_occurrence, last_occurrence)

    string_without_first_last_star = string.replace('*', '', 1).rsplit('*', 1)[0]
    print("–°—Ç—Ä–æ–∫–∞ –±–µ–∑ –ø–µ—Ä–≤–æ–≥–æ –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —Å–∏–º–≤–æ–ª–æ–≤ *:", string_without_first_last_star)
````

### 8 –î–∞–Ω–∞ —Å—Ç—Ä–æ–∫–∞, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ —Å–ª–æ–≤, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã—Ö –ø—Ä–æ–±–µ–ª–∞–º–∏. –í —ç—Ç–æ–π —Å—Ç—Ä–æ–∫–µ:
### —É–¥–∞–ª–∏—Ç–µ –≤—Å–µ –ª–∏—à–Ω–∏–µ –ø—Ä–æ–±–µ–ª—ã (–≤ –Ω–∞—á–∞–ª–µ, –≤ –∫–æ–Ω—Ü–µ, –º–µ–∂–¥—É —Å–ª–æ–≤–∞–º–∏ –æ—Å—Ç–∞–≤–∏—Ç—å —Ä–æ–≤–Ω–æ –æ–¥–∏–Ω –ø—Ä–æ–±–µ–ª);
## –ø–æ–º–µ–Ω—è–π—Ç–µ —Ä–µ–≥–∏—Å—Ç—Ä —Å–∏–º–≤–æ–ª–æ–≤ (—Å—Ç—Ä–æ—á–Ω—ã–µ —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–ø–∏—Å–Ω—ã–º–∏, –ø—Ä–æ–ø–∏—Å–Ω—ã–µ ‚Äì —Å—Ç—Ä–æ—á–Ω—ã–º–∏);
### –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —Å–∫–æ–ª—å–∫–æ –≤ –Ω–µ–π —Å–ª–æ–≤.

````python
string = input("–í–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫—É.  ")
processed_string = ' '.join(string.split()).swapcase()
print("–û–±—Ä–∞–±–æ—Ç–∞–Ω–Ω–∞—è —Å—Ç—Ä–æ–∫–∞:", processed_string)

word_count = len(processed_string.split())
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤ –≤ —Å—Ç—Ä–æ–∫–µ:", word_count)
````

### 9 –°—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ñ–∞–º–∏–ª–∏—é, –∏–º—è –∏ –æ—Ç—á–µ—Å—Ç–≤–æ, –∑–∞–ø–∏—Å–∞–Ω–Ω—ã–µ —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª—ã. –ù–∞–ø—Ä–∏–º–µ—Ä ¬´–ò–≤–∞–Ω–æ–≤ –ò–≤–∞–Ω –ò–≤–∞–Ω–æ–≤–∏—á¬ª. –î–ª—è —ç—Ç–æ–π —Å—Ç—Ä–æ–∫–∏:
### –≤—ã–≤–µ–¥–∏—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –≤ –≤–∏–¥–µ:
### –§–∞–º–∏–ª–∏—è       –ò–≤–∞–Ω–æ–≤
### –ò–º—è           –ò–≤–∞–Ω
### –û—Ç—á–µ—Å—Ç–≤–æ      –ò–≤–∞–Ω–æ–≤–∏—á
### –ø–æ–ª—É—á–∏—Ç–µ —Å—Ç—Ä–æ–∫–∏ –≤–∏–¥–∞ ¬´–ò–≤–∞–Ω–æ–≤ –ò.–ò.¬ª –∏ ¬´–ò.–ò. –ò–≤–∞–Ω–æ–≤¬ª.

````python
full_name = input("–í–≤–µ–¥–∏—Ç–µ –§–∞–º–∏–ª–∏—é, –ò–º—è, –û—Ç—á–µ—Å—Ç–≤–æ")

last_name, first_name, patronymic = full_name.split()

print("–§–∞–º–∏–ª–∏—è\t\t", last_name)
print("–ò–º—è\t\t", first_name)
print("–û—Ç—á–µ—Å—Ç–≤–æ\t", patronymic)

short_name1 = f"{last_name} {first_name[0]}.{patronymic[0]}."
print("–ö—Ä–∞—Ç–∫–æ–µ –∏–º—è 1:\t", short_name1)

short_name2 = f"{first_name[0]}.{patronymic[0]}. {last_name}"
print("–ö—Ä–∞—Ç–∫–æ–µ –∏–º—è 2:\t", short_name2)
````

### 11 –î–ª—è –¥–≤—É—Ö –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã—Ö —Å—Ç—Ä–æ–∫ –ø—Ä–æ–≤–µ—Å—Ç–∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –Ω–∞ —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ —Å–æ–¥–µ—Ä–∂–∏–º–æ–≥–æ –±–µ–∑ —É—á–µ—Ç–∞ —Ä–µ–≥–∏—Å—Ç—Ä–∞ –±—É–∫–≤ –∏ –Ω–∞—á–∞–ª—å–Ω—ã—Ö –∏ –∫–æ–Ω–µ—á–Ω—ã—Ö –ø—Ä–æ–±–µ–ª—å–Ω—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤. –ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è –¥–≤—É—Ö —Å—Ç—Ä–æ–∫ ' HeLLO WOrlD' –∏ 'hello WORLD ' —Ç–∞–∫–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–ª–∂–Ω–æ –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å True (—Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ —Å—Ç—Ä–æ–∫ —Å–æ–≤–ø–∞–¥–∞–µ—Ç).

````python
def compare_strings(str1,str2):
    str1=str1.lower().strip()
    str2=str2.lower().strip()
    
    return str1==str2
string1 = input("–í–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫—É ")
string2 = input("–í–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫—É ")

result = compare_strings(string1,string2)
print(result)
````

### 12 –ü–æ –Ω–æ–º–µ—Ä—É –≥–æ–¥–∞ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –¥–∞–Ω–Ω—ã–π –≥–æ–¥ –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º. (–≥–æ–¥ —è–≤–ª—è–µ—Ç—Å—è –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º, –µ—Å–ª–∏ –µ–≥–æ –Ω–æ–º–µ—Ä –∫—Ä–∞—Ç–µ–Ω 4, –Ω–æ –Ω–µ –∫—Ä–∞—Ç–µ–Ω 100, –∞ —Ç–∞–∫–∂–µ –µ—Å–ª–∏ –æ–Ω –∫—Ä–∞—Ç–µ–Ω 400).

````python
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

year = int(input("–í–≤–µ–¥–∏—Ç–µ –≥–æ–¥: "))

if is_leap_year(year):
    print(f"{year} –≥–æ–¥ —è–≤–ª—è–µ—Ç—Å—è –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º.")
else:
    print(f"{year} –≥–æ–¥ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º.")
````

### 13 –î–∞–Ω—ã —Ç–æ—á–∫–∏  ùê¥=(1,1),ùêµ=(5,5),ùê∂=(1,4),ùê∑=(7,2) . –ù–∞–π—Ç–∏ —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –æ—Ç—Ä–µ–∑–∫–æ–≤  ùê¥ùêµ  –∏  ùê∂ùê∑ .

````python
import math
def find_intersection_point(A, B, C, D):
    x1, y1 = A
    x2, y2 = B
    x3, y3 = C
    x4, y4 = D
    
    intersection_x = ((x1 * y2 - y1 * x2) * (x3 - x4) - (x1 - x2) * (x3 * y4 - y3 * x4)) / ((x1 - x2) * (y3 - y4) - (y1 - y2) * (x3 - x4))
    intersection_y = ((x1 * y2 - y1 * x2) * (y3 - y4) - (y1 - y2) * (x3 * y4 - y3 * x4)) / ((x1 - x2) * (y3 - y4) - (y1 - y2) * (x3 - x4))
    
    return intersection_x, intersection_y

A = (1, 1)
B = (5, 5)
C = (1, 4)
D = (7, 2)

intersection_point = find_intersection_point(A, B, C, D)
print("–¢–æ—á–∫–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è:", intersection_point)
````
