## №10 Имеются два  𝑛 –мерных вектора  𝑥  и  𝑦 , которые задают координаты  𝑛  точек на плоскости (случайные целые числа). Найти наиболее близкие друг другу точки.

````python
import numpy as np

n = int(input("Введите размерность векторов (n): "))
num_points = int(input("Введите количество точек: "))

x = np.random.randint(0, 100, size=(num_points, n))
y = np.random.randint(0, 100, size=(num_points, n))

def euclidean_distance(point1, point2):
    return np.linalg.norm(point1 - point2)

min_distance = float('inf')
closest_points = None

for i in range(num_points):
    for j in range(num_points):
        distance = euclidean_distance(x[i], y[j])
        if distance < min_distance:
            min_distance = distance
            closest_points = (x[i], y[j])

print("Наиболее близкие точки:")
print("Точка x:", closest_points[0])
print("Точка y:", closest_points[1])
print("Расстояние:", min_distance)
````

## №11Треугольник задан координатами своих вершин:  (𝑥1,𝑦1),(𝑥2,𝑦2),(𝑥3,𝑦3) . Значения координат определите с помощью присваивания. Они могут быть нецелыми. Найти периметр и площадь треугольника.


````python
import math

x1, y1 = map(float, input("Введите x1 и y1 через пробел: ").split())
x2, y2 = map(float, input("Введите x2 и y2 через пробел: ").split())
x3, y3 = map(float, input("Введите x3 и y3 через пробел: ").split())

side1 = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
side2 = math.sqrt((x3 - x2)**2 + (y3 - y2)**2)
side3 = math.sqrt((x1 - x3)**2 + (y1 - y3)**2)

P = side1 + side2 + side3
s = P / 2
area = math.sqrt(s*(s-side1)*(s-side2)*(s-side3))

print(f"Периметр : {P:.3f}")
print(f"Площадь : {area:.3f}")
````

## №12 Пусть пользователь вводит 5 чисел: а, b, c, d, e. Реализуйте программу расчета выражения вида  𝑟𝑒𝑠=𝑎+𝑏𝑐2𝑑−𝑒+𝑂𝑠𝑡[𝑏𝑒/𝑐] , где  𝑂𝑠𝑡  - остаток от деления. Учесть невозможность деления на 0.

````python
import math
a = int(input("Введите число a: "))
b = int(input("Введите число b: "))
c = int(input("Введите число c: "))
d = int(input("Введите число d: "))
e = int(input("Введите число e: "))

if c == 0:
    print("Деление на 0 невозможно")
else:
    fraction_result = b * e / c
    remainder = fraction_result % 1
    result = a + (b * c) / (2 * d) - e + remainder
    print (f"Релузльтат : {result:.3f}")
````

````python
string = input("Введите строку:  ")

if '*' not in string:
    print("нет символа")
else:
    first_occurrence = string.find('*')
    last_occurrence = string.rfind('*')

    if first_occurrence == last_occurrence:
        print("Индекс символа *:", first_occurrence)
    else:
        second_occurrence = string.find('*', first_occurrence + 1)
        print("Индексы символа *:", first_occurrence, second_occurrence, last_occurrence)

    string_without_first_last_star = string.replace('*', '', 1).rsplit('*', 1)[0]
    print("Строка без первого и последнего символов *:", string_without_first_last_star)
````

### 12 По номеру года определите, является ли данный год високосным. (год является високосным, если его номер кратен 4, но не кратен 100, а также если он кратен 400).

````python
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

year = int(input("Введите год: "))

if is_leap_year(year):
    print(f"{year} год является високосным.")
else:
    print(f"{year} год не является високосным.")
````

### 13 Даны точки  𝐴=(1,1),𝐵=(5,5),𝐶=(1,4),𝐷=(7,2) . Найти точку пересечения отрезков  𝐴𝐵  и  𝐶𝐷 .

````python
import math
def find_intersection_point(A, B, C, D):
    x1, y1 = A
    x2, y2 = B
    x3, y3 = C
    x4, y4 = D
    
    intersection_x = ((x1 * y2 - y1 * x2) * (x3 - x4) - (x1 - x2) * (x3 * y4 - y3 * x4)) / ((x1 - x2) * (y3 - y4) - (y1 - y2) * (x3 - x4))
    intersection_y = ((x1 * y2 - y1 * x2) * (y3 - y4) - (y1 - y2) * (x3 * y4 - y3 * x4)) / ((x1 - x2) * (y3 - y4) - (y1 - y2) * (x3 - x4))
    
    return intersection_x, intersection_y

A = (1, 1)
B = (5, 5)
C = (1, 4)
D = (7, 2)

intersection_point = find_intersection_point(A, B, C, D)
print("Точка пересечения:", intersection_point)
````
